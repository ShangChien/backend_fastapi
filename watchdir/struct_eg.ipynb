```python
import pandas as pd
target_col_prefix='TARGET'
if 'target' in data:
    label = np.array(data['target'])
    print(label)
    if len(label.shape)==1 or label.shape[1] == 1:
        data[target_col_prefix] = label.reshape(-1)
    else:
        for i in range(label.shape[1]):
            data[target_col_prefix + str(i)] = label[:,i]        
data = pd.DataFrame(data).rename(columns={'atoms': 'SMILES'})
if 'target' in data:
    data = data.drop(columns=['target'])
```

```python
import json
from rdkit import Chem
from rdkit.Chem import rdDepictor
from rdkit.Chem import rdDistGeom
rdDepictor.SetPreferCoordGen(True)
etkdg = rdDistGeom.ETKDGv3()
etkdg.randomSeed = 0xa700f
etkdg.verbose = False
etkdg.numThreads = 0
etkdg.optimizerForceTol = 0.0135
etkdg.useRandomCoords = False
with open('./data/data_Tg.json','r') as f:
    data_f = f.read()
data_raw = json.loads(data_f)
```

```python
data={'smiles':[i['SMILES'] for i in data_raw],'tg':[j['Tg(K)'] for j in data_raw]}
```

```python
structures={'atoms':[],'coordinates':[],'target':[]}
```

```python
for k,v in enumerate(data['smiles']):
    try:
        mol = Chem.MolFromSmiles(v)
        mol = Chem.AddHs(mol)
        cid = rdDistGeom.EmbedMultipleConfs(mol, numConfs=10, params=etkdg)
        symbols= [i.GetSymbol() for i in mol.GetAtoms()]
        for i in cid:
            conf=mol.GetConformer(i)
            positions=conf.GetPositions()
            structures['atoms'].append(symbols)
            structures['coordinates'].append(positions)
            structures['target'].append(data['tg'][k])
    except Exception:
        print('error',v)
```

```python

import pickle
with open("tg_data.pkl", "wb") as file:
    pickle.dump(structures, file)
```
